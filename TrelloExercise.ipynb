```python
import pandas as pd
import numpy as np
import re
from faker import Faker
from faker.providers import DynamicProvider

f = Faker(['in'])
p = Faker(['en'])
```

```python
digital_interest= DynamicProvider(
    provider_name='digital_interest',
    elements=['Travel', 'Food', 'Technology', 'Gaming', 'Music', 'Sport', 'Fashion', 'Movie']
)
f.add_provider(digital_interest)
```
